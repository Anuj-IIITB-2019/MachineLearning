```python
import numpy as np
import pandas as pd

data = pd.read_csv("rainfall.csv")
data
```

```python
data.dropna(how='any', inplace=True)
data
```

```python
months = list(data.columns)[2:14]
months
```

```python
labels = ['Q1','Q2','Q3','Q4']    
for i in range(0,4):   
    j=i*3
    data[labels[i]] = data[months[j]]+data[months[j+1]]+data[months[j+2]]
```

```python
print(list(data.columns))
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data.groupby('SUBDIVISION').mean().sort_values(by='ANNUAL', ascending=False)['ANNUAL']
```

```python
data.groupby('YEAR').sum()['ANNUAL']
```

```python
df2 = data[[months[0],months[1],months[2],months[3]]]
#df2 = df['YEAR','JAN','FEB','MAR','APR']
df2.columns = np.array(['x1','x2','x3','x4'])
for k in range(1,9):
    df3 = data[[months[k],months[k+1],months[k+2],months[k+3]]]
    df3.columns = np.array(['x1','x2','x3','x4'])
    df2 = df2.append(df3)
df2.index = range(df2.shape[0])
    
#df2 = pd.concat([df2, pd.get_dummies(df2['SUBDIVISION'])], axis=1)

# df2.drop('SUBDIVISION', axis=1,inplace=True)
```

```python
msk = np.random.rand(len(df2)) < 0.7

df_train = df2[msk]
df_test = df2[~msk]
df_train.index = range(df_train.shape[0])
df_test.index = range(df_test.shape[0])
```

```python
from sklearn import metrics
reg = linear_model.LinearRegression()
reg.fit(df_train.drop('x4',axis=1),df_train['x4'])
predicted_values = reg.predict(df_train.drop('x4',axis=1))
residuals = predicted_values-df_train['x4'].values
print('Mean Adbsolute Deviation (Training Data): ' + str(np.mean(np.abs(residuals))))
# print("ROOT MEAN SQUARE ERROR  " +str(np.sqrt(metrics.mean_squared_error(residuals,df_train['x4']))))
df_res = pd.DataFrame(residuals)
df_res.columns = ['Residuals']
```

```python
predicted_values = reg.predict(df_test.drop('x4',axis=1))
residuals = predicted_values-df_test['x4'].values
print('Mean Adbsolute Deviation (Test Data): ' + str(np.mean(np.abs(residuals))))
df_res = pd.DataFrame(residuals)
df_res.columns = ['Residuals']
```
